El tiempo entre llegadas de piezas a una estación de ensamblado sigue una distribución exponencial con media 2 minutos. La duración de la tarea llevada a cabo por el operario en dicha estación adopta una distribución exponencial con media 1 minuto por pieza. Determine el tiempo medio de permanencia de las piezas en dicho proceso.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t


def tiempo_permanencia():
    tiempo_llegada = np.random.exponential(scale=2)  # Generar tiempo de llegada de una pieza
    tiempo_tarea = np.random.exponential(scale=1)  # Generar tiempo de duración de la tarea

    tiempo_permanencia = tiempo_llegada + tiempo_tarea  # Calcular tiempo de permanencia

    return tiempo_permanencia

# Realizar una corrida de simulación para determinar el tiempo medio de permanencia
num_piezas = 10000  # Número de piezas a simular
tiempos_permanencia = []

for _ in range(num_piezas):
    tiempo_permanencia_pieza = tiempo_permanencia()
    tiempos_permanencia.append(tiempo_permanencia_pieza)

tiempo_medio_permanencia = np.mean(tiempos_permanencia)

print("Tiempo medio de permanencia:", tiempo_medio_permanencia)
```

```python
# Realizar múltiples simulaciones para determinar el tiempo medio de permanencia promedio
num_simulaciones = 1000  # Número de simulaciones a realizar
num_piezas = 100  # Número de piezas por simulación
tiempos_medios_permanencia = []

for _ in range(num_simulaciones):
    tiempos_permanencia = []

    for _ in range(num_piezas):
        tiempo_permanencia_pieza = tiempo_permanencia()
        tiempos_permanencia.append(tiempo_permanencia_pieza)

    tiempo_medio_permanencia = np.mean(tiempos_permanencia)
    tiempos_medios_permanencia.append(tiempo_medio_permanencia)

tiempo_medio_permanencia_promedio = np.mean(tiempos_medios_permanencia)

print("Tiempo medio de permanencia promedio:", tiempo_medio_permanencia_promedio)
```

```python
# Graficar la estabilidad de las simulaciones
plt.plot(tiempos_medios_permanencia, color='red')
plt.xlabel('Simulación')
plt.ylabel('Tiempo medio de permanencia')
plt.title('Estabilidad del tiempo medio de permanencia en el proceso')
plt.show()
```

```python
# Graficar un histograma del tiempo medio de permanencia
plt.hist(tiempos_medios_permanencia, bins=30, color='red', edgecolor='orange')
plt.xlabel('Tiempo medio de permanencia')
plt.ylabel('Frecuencia')
plt.title('Histograma del tiempo medio de permanencia en el proceso')
plt.show()
```

```python
# Calcular la media y desviación estándar del tiempo medio de permanencia
media_tiempo_medio_permanencia = np.mean(tiempos_medios_permanencia)
desviacion_tiempo_medio_permanencia = np.std(tiempos_medios_permanencia)

print("Media del tiempo medio de permanencia:", media_tiempo_medio_permanencia)
print("Desviación estándar del tiempo medio de permanencia:", desviacion_tiempo_medio_permanencia)
```

```python
# Calcular el intervalo de confianza del tiempo medio de permanencia
grados_libertad = num_simulaciones - 1
media_tiempo_medio_permanencia = np.mean(tiempos_medios_permanencia)
desviacion_tiempo_medio_permanencia = np.std(tiempos_medios_permanencia)
intervalo_confianza = t.interval(0.95, df=grados_libertad, loc=media_tiempo_medio_permanencia, scale=desviacion_tiempo_medio_permanencia / np.sqrt(num_simulaciones))

print("Intervalo de confianza (95%):", intervalo_confianza)
```
