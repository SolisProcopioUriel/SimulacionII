**Ejercicio 19:** Las bombillas eléctricas de un fabricante $A$ tiene una duración media de 1400 horas con una desviación típica de 200 horas, mientras que las de otro fabricante $B$ tiene una duración media de 1200 horas con una desviación típica de 100 horas. Si se toman mustras al azar de 125 bombillas de cada fabricante, ¿Cuál es la probabilidad de que las bombillas de $A$ tengan una duración media que sea al menos (a) 60 horas, (b) 250 horas más que las bombillas de $B$?

#**Simulación** Para el inciso (a)

```python
import numpy as np
from scipy.stats import norm
import math
import random
import matplotlib.pyplot as plt
```

```python
# Función para simular el tiempo de duración de las bombillas de un fabricante
def duracion_bombillas(media, desviacion, n):
    muestra = [random.normalvariate(media, desviacion) for i in range(n)] # Generar una muestra aleatoria de n bombillas con la distribución normal dada
    duracion_media = sum(muestra) / n  # Calcular la duración media de la muestra
    return duracion_media
```

```python
# Función para simular una corrida de la prueba de duración de bombillas
def simulacion_1(media_A, desviacion_A, media_B, desviacion_B, n):
    duracion_A = duracion_bombillas(media_A, desviacion_A, n) # Simular la duración media de las bombillas del fabricante A    
    duracion_B = duracion_bombillas(media_B, desviacion_B, n) # Simular la duración media de las bombillas del fabricante B
    
    diferencia = duracion_A - duracion_B # Calcular la diferencia de duración media entre las dos muestras
    return diferencia
```

```python
simulacion_1(1400,200,1200,100,125)
```

```python
def simulacion(n):
    L = []# Lista vacía que contendrá las diferencias de duración media de las bombillas de A y B en cada simulación
    prob_final = [] # Lista vacía que contendrá las probabilidades de que la duración media de las bombillas de A sea al menos 160 horas mayor que la de B
    for i in range(n): # Iteramos n veces para hacer n simulaciones
        duracion_bombillas_A = np.random.normal(1400, 200, 125)  # Generamos una muestra aleatoria de duración de bombillas A con media 1400 y desviación 200
        duracion_bombillas_B = np.random.normal(1200, 100, 125) # Generamos una muestra aleatoria de duración de bombillas B con media 1200 y desviación 100
        diferencia_medias = duracion_bombillas_A.mean() - duracion_bombillas_B.mean() # Calculamos la diferencia de duración media de las dos muestras
        L.append(diferencia_medias)
        prob_final.append((np.array(L) >= 160).mean()) # Calculamos la probabilidad de que la duración media de las bombillas A sea al menos 160 horas mayor que la de B y la añadimos a la lista prob_final
    return prob_final
```

```python
simulacion(300)
```

```python
X = simulacion(1000) #hacemos 1000 veces la simuación
```

```python
plt.plot(X, color = "red") #Gráficar
plt.xlabel('iteraciones') #Etiqueta del eje de las x
plt.ylabel('Probabilidad') #Etiqueta del eje de las y
plt.show() #Mostrar la gráfica
```

```python
np.mean(X) #calculamos la media
```

```python
#Repetimos el modelo n veces
def modelo(n):
    lista=[]
    for i in range(n):
        s = simulacion(1000) # Realizamos la para calcualar 1000 veces
        lista.append(np.mean(s)) # Agregamos la media
    return lista
```

```python
Y=modelo(200) # Asignamos valores a n 
```

```python
plt.hist(Y, color = "red") #Gráficamos un histograma 
plt.show()#mostramos la gráfica
```

```python
print(np.mean(Y), np.std(Y)) #imprimimos la media y la desviación estandar
```

```python
np.mean(Y) - 1.96*np.std(Y)/np.sqrt(50), np.mean(Y) + 1.96*np.std(Y)/np.sqrt(50) #Calculamos el intervalo de confinza
```
