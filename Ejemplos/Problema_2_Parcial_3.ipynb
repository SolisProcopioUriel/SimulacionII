La central de un servicio de emergencias dispone de tres teléfonos para atender las llamadas. La duración de las llamadas sigue una distribución normal con los parámetros que recoge la tabla en minutos

| Unidad vendidas | Media | Desviación |
|------           |-----|-----|
|1 |  2.0  |0.4   |
|2 | 2.1   | 0.3  |
|3 | 1.9   | 0.5  |

Determinar el tiempo de respuesta de la central de emergencias.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t


def tiempo_respuesta():
    unidades = [1, 2, 3]  # Números de las unidades
    medias = [2.0, 2.1, 1.9]  # Medias de las distribuciones normales de duración
    desviaciones = [0.4, 0.3, 0.5]  # Desviaciones estándar de las distribuciones normales de duración

    duraciones = []  # Lista para almacenar las duraciones de las llamadas

    for unidad, media, desviacion in zip(unidades, medias, desviaciones):
        duracion = np.random.normal(loc=media, scale=desviacion)  # Generar una duración aleatoria para la unidad
        duraciones.append(duracion)

    tiempo_respuesta = max(duraciones)  # El tiempo de respuesta será el máximo de las duraciones de las llamadas

    return tiempo_respuesta

# Realizar una corrida de simulación para determinar el tiempo de respuesta
tiempo_respuesta_simulado = tiempo_respuesta()

print("Tiempo de respuesta simulado:", tiempo_respuesta_simulado)
```

```python
# Realizar múltiples simulaciones del tiempo de respuesta
num_simulaciones = 10000
tiempos_respuesta_simulados = []

for _ in range(num_simulaciones):
    tiempo_respuesta_simulado = tiempo_respuesta()
    tiempos_respuesta_simulados.append(tiempo_respuesta_simulado)

# Calcular la media y desviación estándar de los tiempos de respuesta simulados
media_tiempo_respuesta = np.mean(tiempos_respuesta_simulados)
desviacion_tiempo_respuesta = np.std(tiempos_respuesta_simulados)

print("Tiempo de respuesta promedio:", media_tiempo_respuesta)
print("Desviación estándar del tiempo de respuesta:", desviacion_tiempo_respuesta)
```

```python
# Graficar la estabilidad de las simulaciones
plt.plot(tiempos_respuesta_simulados, color='red')
plt.xlabel('Simulación')
plt.ylabel('Tiempo de respuesta')
plt.title('Estabilidad del tiempo de respuesta en 10000 simulaciones')
plt.show()
```

```python
# Graficar un histograma de los tiempos de respuesta simulados
plt.hist(tiempos_respuesta_simulados, bins=30, color='red', edgecolor='orange')
plt.xlabel('Tiempo de respuesta')
plt.ylabel('Frecuencia')
plt.title('Histograma del tiempo de respuesta en 10000 simulaciones')
plt.show()
```

```python
# Calcular la media y desviación estándar de los tiempos de respuesta simulados
media_tiempo_respuesta = np.mean(tiempos_respuesta_simulados)
desviacion_tiempo_respuesta = np.std(tiempos_respuesta_simulados)

print("Media del tiempo de respuesta:", media_tiempo_respuesta)
print("Desviación estándar del tiempo de respuesta:", desviacion_tiempo_respuesta)
```

```python
# Calcular el intervalo de confianza del tiempo de respuesta
grados_libertad = num_simulaciones - 1
media_tiempo_respuesta = np.mean(tiempos_respuesta_simulados)
desviacion_tiempo_respuesta = np.std(tiempos_respuesta_simulados)
intervalo_confianza = t.interval(0.95, df=grados_libertad, loc=media_tiempo_respuesta, scale=desviacion_tiempo_respuesta / np.sqrt(num_simulaciones))

print("Intervalo de confianza (95%):", intervalo_confianza)
```
