```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 15 números aleatorios
random_numbers = np.random.rand(15)

# Generar números con distribución exponencial
exponential_numbers = np.random.exponential(scale=1.0, size=1000)

# Crear una figura y subtramas
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Graficar los números aleatorios
axes[0].hist(random_numbers, bins=15, color='orange', alpha=0.7)
axes[0].set_title('Números Aleatorios')
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Frecuencia')

# Graficar la distribución exponencial
axes[1].hist(exponential_numbers, bins=30, color='red', alpha=0.7)
axes[1].set_title('Distribución Exponencial')
axes[1].set_xlabel('Valor')
axes[1].set_ylabel('Frecuencia')

# Ajustar el espaciado entre subtramas y mostrar la figura
plt.tight_layout()
plt.show()
```
